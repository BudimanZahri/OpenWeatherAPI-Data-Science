```python
# Budiman Zahri
# Teknik Mekatronika

import requests
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
from datetime import datetime, timedelta
from api_key import API_KEY
```

```python
cities = ['Singapore', 'Perth', 'Jakarta', 'Brisbane', 'Melbourne', 'Sydney', 'Batam']
weather_data = []
```

```python
error_messages = {
    'en': f"Error for {{city}}: {{code}}",
    'id': f"Error di {{city}}: {{code}}",
    'zh': f"城市{{city}}出错: {{code}}"
}

for city in cities:
    response = requests.get(f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric')
    # print(response.json().dumps(data, indent=4),'res')
    # console.log(response,'ambatukam');
    if response.ok:
        data = response.json()
        print(json.dumps(data, indent=4)) 
        local_time = datetime.utcfromtimestamp(data['dt']) + timedelta(hours=7)

        weather_data.append({
            'City': data['name'],
            'Temperature (°C)': data['main']['temp'],
            'Feels Like (°C)': data['main']['feels_like'],
            'Humidity (%)': data['main']['humidity'],
            'Weather': data['weather'][0]['description'],
            'Wind Speed (m/s)': data['wind']['speed'],
            'Time (UTC+7)': local_time.strftime('%d-%m-%Y %H:%M:%S')
        })
    else:
        print(error_messages['en'].format(city=city, code=response.status_code))
        print(error_messages['id'].format(city=city, code=response.status_code))
        print(error_messages['zh'].format(city=city, code=response.status_code))
```

```python
df = pd.DataFrame(weather_data)
df.to_csv('weather_data.csv', index=False)
df.head()
```

```python
average_temp = df['Temperature (°C)'].mean()
highest_temp_city = df.loc[df['Temperature (°C)'].idxmax(), ['City', 'Temperature (°C)']]
lowest_temp_city = df.loc[df['Temperature (°C)'].idxmin(), ['City', 'Temperature (°C)']]
common_weather = df['Weather'].mode()[0]

print(f"Average Temperature: {average_temp:.2f} °C")
print(f"City with Highest Temperature: {highest_temp_city['City']} ({highest_temp_city['Temperature (°C)']} °C)")
print(f"City with Lowest Temperature: {lowest_temp_city['City']} ({lowest_temp_city['Temperature (°C)']} °C)")
print(f"Most Common Weather: {common_weather}")
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='City', y='Temperature (°C)', data=df, hue='City', palette='viridis', dodge=False)
plt.xlabel('City')
plt.ylabel('Temperature (°C)')
plt.title('Temperature in Different Cities')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
